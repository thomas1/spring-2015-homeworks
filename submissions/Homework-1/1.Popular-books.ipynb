## Step 2. Parsing the responses

In this section, we practice some of the basic Python tools that we learned in class and the powerful string handling methods that Python offers. Our goal is to be able to pick the interesting parts of the response and transform them in a format that will be useful to us.

Our first task will be to isolate the names of all the best-seller lists of the NYT. Fill in the rest of the ```print_names_from_XML()``` function that reads the XML response and prints all these names. **(5 pts)** 

Hint: Our _pretty_ formatter puts each tag on a separate line. You may want to read the documentation of the `split()`, `strip()` and `startswith()` functions.

````python
import re
def print_names_from_XML(response):
    xml_parser = xml.dom.minidom.parseString(response.text)
    pretty_response = xml_parser.toprettyxml()   
    #Set the start and end conditions
    start= r'<list_name>'
    end= r'</list_name>'
    #add the indices of each start and end value into a list
    list_start = [m.start() for m in re.finditer(start, pretty_response)]
    list_end = [m.end() for m in re.finditer(end, pretty_response)]
    #print the values between the 2 index values
    for x in range (1, len(list_start)):
            print pretty_response[list_start[x]+len(start):list_end[x]-len(end)]
    return 

    
````

````python
response = requests.get("http://api.nytimes.com/svc/books/v3/lists/names.xml?api-key=%s"%(key))
print_names_from_XML(response)
````

Can you do the same thing for the JSON response? Notice that a JSON object is basically a dictionary. **(5 pts)**

````python
def print_names_from_JSON(response):
    #uses same convention as above for a cleaner list to work with
    pretty_response = json.dumps(response.json())
    
    start= r'display_name": "'
    end= r'", "list_name_encoded"'
    
    list_start = [m.start() for m in re.finditer(start, pretty_response)]
    list_end = [m.end() for m in re.finditer(end, pretty_response)]
    
    for x in range (1, len(list_start)):
            print pretty_response[list_start[x]+len(start):list_end[x]-len(end)]
    return 
````

````python
response = requests.get("http://api.nytimes.com/svc/books/v3/lists/names.json?api-key=%s"%(key))
print_names_from_JSON(response)
````

Write a function that, given a list name and a date, returns a tuple with the books that were best-sellers for that date and the date on which the list was published by the NYT. **(40 pts)**

````python
import datetime
import time

def get_books(date, list_name):
    books_set = []
    published_date = []
    
    response = requests.get("http://api.nytimes.com/svc/books/v3/lists/%s/%s.json?&api-key=%s"%(date,list_name,key))
    pretty_response = json.dumps(response.json(), indent=2)
    
    #Gets the number of results from the json dump
    num_start = r'"num_results": '
    num_end = r'"results":'
    start_numresults = [m.start() for m in re.finditer(num_start, pretty_response)]
    end_numresults = [m.end() for m in re.finditer(num_end, pretty_response)]
    num_results = pretty_response[start_numresults[0]+len(num_start):end_numresults[0]-len(num_end)-5]
    print "Number of Results: " + str(int(num_results))
    
    #Resets pretty_response into a null string. Repopulates the string by full traversing all pages in increments of 20. 
    pretty_response = ""
    for x in range (0,int(num_results),20):
        offset = x
        response = requests.get("http://api.nytimes.com/svc/books/v3/lists/%s/%s.json?&offset=%s&api-key=%s"%(date,list_name,offset,key))
        pretty_response = pretty_response + json.dumps(response.json(), indent=2)
    
    #Find the title index of each entry and append it to the books_set list
    start_title = r'"title": '
    end_title = r'"contributor_note"'
    list_start_title = [m.start() for m in re.finditer(start_title, pretty_response)]
    list_end_title = [m.end() for m in re.finditer(end_title, pretty_response)]
    
    for x in range (0, len(list_start_title)):
            books_set.append(pretty_response[list_start_title[x]+len(start_title)+1:list_end_title[x]-len(end_title)-12])
    
    #Find the best seller date of each entry and append it to the booksellers_date list
    start = r'"bestsellers_date"'
    end = r'"books"'
    list_start = [m.start() for m in re.finditer(start, pretty_response)]
    list_end = [m.end() for m in re.finditer(end, pretty_response)]
    published_date.append(pretty_response[list_start[0]+len(start)+3:list_end[0]-len(end)-8])


    return (books_set,published_date)

date = datetime.date(2014,7,1)
list_name = "hardcover-fiction"
get_books(date, list_name)
````

Notice that the free API key that we have has a limit of 8 QPS (queries per second). If we send multiple queries and pass this limit, we will get back an error instead of the answer. To avoid this situation, a naive way is, after each query, to wait $1/8=0.125$ seconds. The command for this is 
```python
time.sleep(0.125)
```

Great! The final step is to write a function that takes a time window and a list name, and returns a dictionary with the books that were best-sellers as keys and the number of weeks that they were in the list as values. **(30 pts)**

````python
import datetime

def most_popular(start_date, end_date, list_name):
    books_set = []
    weeks_on_list = []
    date = start_date
    response = requests.get("http://api.nytimes.com/svc/books/v3/lists/%s/%s.json?weeks-on-list?&api-key=%s"%(date,list_name,key))
    pretty_response = json.dumps(response.json(), indent=2)
    
    #Gets the number of results from the json dump
    num_start = r'"num_results": '
    num_end = r'"results":'
    start_numresults = [m.start() for m in re.finditer(num_start, pretty_response)]
    end_numresults = [m.end() for m in re.finditer(num_end, pretty_response)]
    num_results = pretty_response[start_numresults[0]+len(num_start):end_numresults[0]-len(num_end)-5]
    print "Number of Results: " + str(int(num_results))
    
    #Resets pretty_response into a null string. Repopulates the string by full traversing all pages in increments of 20. 
    pretty_response = ""
    for x in range (0,int(num_results),20):
        offset = x
        response = requests.get("http://api.nytimes.com/svc/books/v3/lists/%s/%s.json?&offset=%s&api-key=%s"%(date,list_name,offset,key))
        pretty_response = pretty_response + json.dumps(response.json(), indent=2)
        
    #Find the title index of each entry and append it to the books_set list
    start_title = r'"title": '
    end_title = r'"contributor_note"'
    list_start_title = [m.start() for m in re.finditer(start_title, pretty_response)]
    list_end_title = [m.end() for m in re.finditer(end_title, pretty_response)]
    
     #Find the weeks on list of each entry and append it to the weeks_on_list list
    start_weeks = r'"weeks_on_list": '
    end_weeks= r'"article_chapter_link":'
    list_start_weeks = [m.start() for m in re.finditer(start_weeks, pretty_response)]
    list_end_weeks = [m.end() for m in re.finditer(end_weeks, pretty_response)]
    
    #Add the books and their number of weeks on the list to the lists
    for x in range (0, len(list_start_title)):
            books_set.append(pretty_response[list_start_title[x]+len(start_title)+1:list_end_title[x]-len(end_title)-12])
            weeks_on_list.append(pretty_response[list_start_weeks[x]+len(start_weeks):list_end_weeks[x]-len(end_weeks)-11])
   
    #Add the book and it's number of weeks on list to the dictionary        
    book_dict = dict([(books_set[0], 1), (books_set[1], 1)])        
    for x in range (0,len(books_set)):
        book_dict.update({books_set[x]:weeks_on_list[x]})

    return book_dict
````
