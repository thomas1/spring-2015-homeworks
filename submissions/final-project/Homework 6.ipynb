```python
carbon_monoxide = df[(df['Parameter Name'] == 'Carbon monoxide') & (df['Sample Duration'] == '1 HOUR')]

state_list = carbon_monoxide['State Name'].tolist()
carbon_list = carbon_monoxide['Observation Count'].tolist()

carbon_largest = heapq.nlargest(10,carbon_list)
carbon_topten = []
for x in carbon_largest:
    carbon_topten.append((x,state_list[carbon_list.index(x)]))

print "\nTop ten emissions of Carbon Monoxide", carbon_topten

ozone = df[(df['Parameter Name'] == 'Ozone') & (df['Sample Duration'] == '1 HOUR')]

state_list = ozone['State Name'].tolist()
ozone_list = ozone['Observation Count'].tolist()

ozone_largest = heapq.nlargest(10,ozone_list)
ozone_topten = []

for x in ozone_largest:
    ozone_topten.append((x,state_list[ozone_list.index(x)]))

print "\nTop ten emissions of Ozone", ozone_topten


nitrogen_dioxide = df[(df['Parameter Name'] == 'Nitrogen dioxide (NO2)') & (df['Sample Duration'] == '1 HOUR')]

state_list = nitrogen_dioxide['State Name'].tolist()
no2_list = nitrogen_dioxide['Observation Count'].tolist()

no2_largest = heapq.nlargest(10,no2_list)
no2_topten = []

for x in no2_largest:
    no2_topten.append((x,state_list[no2_list.index(x)]))

print "\nTop ten emissions of Nitrogen Dioxide", no2_topten


sulfur_dioxide = df[(df['Parameter Name'] == 'Sulfur dioxide') & (df['Sample Duration'] == '1 HOUR')]

state_list = sulfur_dioxide['State Name'].tolist()
so2_list = sulfur_dioxide['Observation Count'].tolist()

so2_largest = heapq.nlargest(10,so2_list)
so2_topten = []

for x in so2_largest:
    so2_topten.append((x,state_list[so2_list.index(x)]))

print "\nTop ten emissions of Sulfur Dioxide", so2_topten
```

```python
df_all = pd.DataFrame(data_all, columns=['State Code','Year','Parameter Name','Sample Duration','Observation Count','Arithmetic Mean',
                                         'Units of Measure','State Name'])
df_all
```

```python
year = 1990

for i in range(0,25):
    carbon_monoxide_all = df_all[(df_all['Parameter Name'] == 'Carbon monoxide') & (df_all['Sample Duration'] == '8-HR RUN AVG END HOUR')
                             & (df_all['State Code'] <= 1000) & (df_all['Year'] == year)]

    ozone_all = df_all[(df_all['Parameter Name'] == 'Ozone') & (df_all['Sample Duration'] == '8-HR RUN AVG END HOUR')
                             & (df_all['State Code'] <= 80)
                              & (df_all['Year'] == year)]
    
    average_co_all = carbon_monoxide_all['Observation Count'].mean()
    co_std_all = numpy.std(carbon_monoxide_all['Observation Count'], axis=0)
    print " The average CO emission across all stations is", year, average_co_all,"parts per million"
    print " The Standard Deviation of CO emission across all stations is",year, co_std_all,"parts per million"

    average_ozone_all = ozone_all['Observation Count'].mean()
    ozone_std_all = numpy.std(ozone_all['Observation Count'], axis=0)
    print " The average Ozone emission across all stations is", year, average_ozone_all,"parts per million"
    print " The Standard Deviation of CO emission across all stations is", year, ozone_std_all,"parts per million"
    
    year +=1

```

```python
year = 1990

carbon_monoxide_all = df_all[(df_all['Parameter Name'] == 'Carbon monoxide') & (df_all['Sample Duration'] == '8-HR RUN AVG END HOUR')
                             & (df_all['State Code'] <= 1000) & (df_all['Year'] == year)]




fig41 = pl.figure()
fig41.suptitle('Carbon Monoxide Chart', fontsize=20)
ax41 = pl.subplot(111)

fig42 = pl.figure()
fig42.suptitle('Ozone Chart', fontsize=20)
ax42 = pl.subplot(111)


fig43 = pl.figure()
fig43.suptitle('NO2 Chart', fontsize=20)
ax43 = pl.subplot(111)

fig44 = pl.figure()
fig44.suptitle('SO2 Chart', fontsize=20)
ax44 = pl.subplot(111)

for i in range(1,25):
    carbon_monoxide_all = df_all[(df_all['Parameter Name'] == 'Carbon monoxide') & (df_all['Sample Duration'] == '8-HR RUN AVG END HOUR')
                             & (df_all['State Code'] <= 1000) & (df_all['Year'] == year)]

    ozone_all = df_all[(df_all['Parameter Name'] == 'Ozone') & (df_all['Sample Duration'] == '1 HOUR')
                             & (df_all['State Code'] <= 1000) & (df_all['Year'] == year)]
    
    no2_all = df_all[(df_all['Parameter Name'] == 'Nitrogen dioxide (NO2)') & (df_all['Sample Duration'] == '1 HOUR')
                             & (df_all['State Code'] <= 1000) & (df_all['Year'] == year)]
    
    so2_all = df_all[(df_all['Parameter Name'] == 'Sulfur dioxide') & (df_all['Sample Duration'] == '1 HOUR')
                             & (df_all['State Code'] <= 1000) & (df_all['Year'] == year)]
    
    ax41.scatter(range(len(carbon_monoxide_all['Year'])),carbon_monoxide_all['Observation Count'],alpha=.1)
    
    ax42.scatter(range(len(ozone_all['Year'])),ozone_all['Observation Count'],alpha=.1)
    
    ax43.scatter(range(len(no2_all['Year'])),no2_all['Observation Count'],alpha=.1)
    
    ax44.scatter(range(len(so2_all['Year'])),so2_all['Observation Count'],alpha=.1)
    
    
    
    year +=1
```
